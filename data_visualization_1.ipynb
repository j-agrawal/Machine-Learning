```
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```
df = open('student.csv','r')
```

```
file = csv.reader(df)
```

```
next(file,None)
```

```
student_name = []
marks = []
age = []
contact = []
study_hours = []
```

```
for row in file :
  student_name.append(row[0])
  marks.append(row[1])
  age.append(row[2])
  contact.append(row[3])
  study_hours.append(row[4])
 
```

```
explode= (0.1, 0, 0, 0)
plt.pie(marks,labels=student_name,explode=explode,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title("marks obtained by students")
plt.show()
```

```
explode= (0.1, 0, 0, 0)
plt.plt(study_hours,labels=student_name,explode=explode,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title("study hours of each student")
plt.show()
```
